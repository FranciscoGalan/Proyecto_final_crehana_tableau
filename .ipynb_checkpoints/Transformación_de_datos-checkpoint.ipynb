# Fusionar datasets 

```python
import pandas as pd
```

```python
data_2 = pd.read_csv("dataset_2.csv")
data_2.shape
```

```python
data_2.head()
```

```python
data_3 = pd.read_csv("dataset_3.csv")
data_3.shape
```

```python
data_3.head()
```

```python
data_3_filtered = data_3[["game_title", 'review']]
```

```python
data_3_filtered.head()
```

```python
data_3_filtered.loc[:, "review"] = round(data_3_filtered.review / 10, 1)
```

```python
data_3_filtered.head()
```

```python
data_3_filtered.columns = ["game_title", "review_other"]
```

```python
data = data_2.merge(data_3_filtered, how='left', left_on='name', right_on='game_title')
data.head()
```

```python
data.columns
```

# Limpiar dataset

```python
data_f = data[['name', 'platform', 'year_of_release', 'genre', 'publisher', 'na_sales',
       'eu_sales', 'jp_sales', 'other_sales', 'critic_score',
       'critic_count', 'user_score', 'user_count', 'review_other', 'rating']].sort_values(by='year_of_release').reset_index(drop=True)
```

```python
data_f.head(10)
```

```python
data_f.shape
```

```python
# Verificar valores nulos
data_f.isna().sum()
```

```python
# Tumbar NAs en columnas
data_ff = data_f.dropna(subset=['name', 'year_of_release', 'publisher'])
data_ff.shape
```

```python
data_ff.isna().sum()
```

```python
# Eliminar años en los que tenemos muy pocos valores
data_ff.year_of_release.value_counts()
```

```python
data_ff = data_ff.loc[~(data_ff.year_of_release < 1981) & ~(data_ff.year_of_release > 2016), :]
```

```python
# Pasar nombres de columnas a español
data_ff.columns
```

```python
data_ff.columns = ['nombre', 'plataforma', 'año_de_lanzamiento', 'género', 'distribuidora', 'ventas_na',
       'ventas_eu', 'ventas_jp', 'ventas_otros', 'critica_puntuacion', 'critica_conteo',
       'usuarios_puntuacion', 'usuarios_conteo', 'puntuacion_otro', 'clasificacion']
```

```python
data_ff.head()
```

```python
data_ff.shape
```

```python
# Insertar una columna de ID para cada juego
data_ff.insert(0, 'ID', list(range(1, data_ff.shape[0]+1)))
```

```python
data_ff.tail()
```
